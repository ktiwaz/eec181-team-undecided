```python
import numpy as np

LUT_SIZE = 313       # LUT index goes from 0 to 312
SHIFT = 9            # Right-shift for LUT indexing (>>9 = /512)
SIGMA = 240.0        # Adjust for vignette smoothness
MAX_WEIGHT = 255     # 8-bit output

lut = []
for i in range(LUT_SIZE):
    d2 = i << SHIFT  # i * 512
    weight = np.exp(-d2 / (2 * SIGMA ** 2))
    value = int(round(weight * MAX_WEIGHT))
    lut.append(min(value, MAX_WEIGHT))

# Output Verilog case statement
with open("vignette_lut_case.v", "w") as f:
    f.write("// Auto-generated vignette LUT case statement\n")
    f.write("always @(*) begin\n")
    f.write("    case (addr)\n")
    for addr, val in enumerate(lut):
        f.write(f"        9'd{addr}: weight = 8'd{val};\n")
    f.write("        default: weight = 8'd0;\n")
    f.write("    endcase\n")
    f.write("end\n")

print("Done: Wrote vignette_lut_case.v with 313 entries.")
```
